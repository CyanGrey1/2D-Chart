```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mpl
import numpy as np
import ipywidgets as ipyw
```

```python
con = sql.connect("overall_data.db")    
cur = con.cursor()  
```

```python
Query = "SELECT DISTINCT Country FROM newdata"
cur.execute(Query)
c = cur.fetchall()
c = np.array(c)
```

```python
def getdata_country(country1, country2, things_to_get):
    things = ', '.join(things_to_get)
    Query=f"SELECT {things} FROM newdata WHERE Country='{country1}' OR Country='{country2}' "  
    cur.execute(Query)
    data= cur.fetchall()  
    return np.array(data).T 
```

```python
def update(countryA,countryB):
    data = getdata_country(countryA, countryB, ['year', 'fossil_fuels', 'population', 'gdp'])


    fig9 = plt.figure(figsize=(10,6))
    ax=fig9.add_subplot()
    ax.clear()             
    sc_plot=ax.scatter(data[0], data[1]/data[2]*1000000, s=60, c=data[3]/data[2], cmap='GnBu',
                       ec='k', linewidth=1)

    ax.set_xlabel(r'Year', fontsize=16)
    ax.set_ylabel(r'Fossil Fuel Consumption per Capita (MWh)', fontsize=15)
    plt.xticks(rotation=65)
    plt.title('Fuel Consumption by Capita by Year', weight='bold', fontsize=16)
    

    cax = fig9.add_axes([0.15, 0.8, 0.2, 0.05])
    fig9.colorbar(plt.cm.ScalarMappable(cmap='GnBu'),
    cax=cax, orientation='horizontal', pad=1100, label='GDP per Capita',)
    fig9.canvas.draw()


form_item_layout = ipyw.widgets.Layout(
    display='flex',
    flex_flow='row',
    justify_content='space-between'
) 


CountryA = ipyw.widgets.Dropdown(options=c[:,0],value='Mexico',description='Country:',disabled=False)
CountryB = ipyw.widgets.Dropdown(options=c[:,0],value='China',description='Country:',disabled=False)


form_items = [
    ipyw.widgets.Box([ipyw.widgets.Label(value='First Country:'),
         CountryA], layout=form_item_layout),
    ipyw.widgets.Box([ipyw.widgets.Label(value='Second Country:'),
         CountryB], layout=form_item_layout)]

ipyw.interact(update, countryA=CountryA, countryB = CountryB);
```
